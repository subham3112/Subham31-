```python
from tkinter import *
import os
import sqlite3 
'''           ------------         GUI (1)      -------------      '''
def loggedin():
    
    global screen2
    global name
    global branch
    global regdid
    global name_entry
    global branch_entry
    global regdid_entry

    screen2 = Toplevel(screen)
    screen2.title("Student Details")
    screen2.geometry("500x500")
    screen2.configure(bg = "aquamarine")
    
    name = StringVar()
    branch = StringVar()
    regdid = StringVar()

    
    Label(screen2, text = "Name :",font = ("arial",10,"bold")).place(x = 40,y = 70)

    name_entry = Entry(screen2,textvariable = name)
    name_entry.place(x = 150,y = 70)
   
    Label(screen2, text = "Branch :",font = ("arial",10,"bold")).place(x = 40,y = 120)

    branch_entry = Entry(screen2,textvariable = branch)
    branch_entry.place(x = 150,y = 120)

   
    Label(screen2, text = "Regd Id :",font = ("arial",10,"bold")).place(x = 40,y = 170)

    regdid_entry = Entry(screen2,textvariable = regdid)
    regdid_entry.place(x = 150,y = 170)

   
    Button(screen2, text = "Continue",fg ='green',width = 14,font = ("arial",10,"bold"), relief = RAISED , command = do).place(x = 180,y = 250)
  
    
'''         ---------------       GUI(2)      --------------        '''
   
def open_window():
    
    tip = Toplevel()
    tip.title("SUBJECTS")
    tip.geometry("500x500")
    tip.configure(bg = "aquamarine")
    
    Label(tip, text = "MARKS SUBMISSION",font = ("arial",14,"bold")).place(x=145,y=40)
    
    Label(tip, text = "Input for respective marks :",font = ("arial",11,"bold")).place(x=50,y=90)
    
    Button(tip, text = "FCT",width = 10,bg = "yellow", command = open_window_fct).place(x=70,y=130)
    
    Button(tip, text = "IDS", width =10,bg = "yellow", command = open_window_ids).place(x=70,y=200)

    Button(tip, text = "EMFT", width = 10,bg = "yellow", command = open_window_emft).place(x=250,y=160)

    Button(tip, text = "SUBMIT",fg = "black",bg = "light blue",font = ("arial",9,"bold"),width =10, command = open_window3).place(x=160,y=270)
 
'''       ---------------      SUBJECTS SUBMISSION        ----------------      '''
'''         ----------           FCT SUBMISSION         ------------     '''
    
def open_window_fct():
    global fct_marks
    global fct_entry
    fct_marks = IntVar()
    
    fct = Toplevel()
    fct.title("FCT")
    fct.geometry("400x400")
    Label(fct, text = "Enter the marks obtained :").pack()
    
    fct_entry = Entry(fct,textvariable = fct_marks)
    fct_entry.pack()
    
    Button(fct, text = "Done", command = fct_submission).pack()
      
def fct_submission():
            
    fct1 = fct_marks.get()
    name1 = name.get()
        
    conn = sqlite3.connect("New.db")
    with conn:
        cur = conn.cursor()
#    addColumn = "ALTER TABLE DETAILS ADD COLUMN FCT varchar(32)"
#    cur.execute(addColumn)
    cur.execute( '''UPDATE DETAILS SET FCT == "%s" WHERE Name == "%s"''' %(fct1,name1))
    conn.commit()
    cur.close()    
    conn.close()
'''    ------------       IDS SUBMISSION      ------------    '''    
    
def open_window_ids():
    global ids_marks
    global ids_entry
    ids_marks = IntVar()
    
    ids = Toplevel()
    ids.title("IDS")
    ids.geometry("400x400")
    Label(ids, text = "Enter the marks obtained :").pack()
    
    ids_entry = Entry(ids,textvariable = ids_marks)
    ids_entry.pack()
    
    Button(ids, text = "Done", command = ids_submission).pack()
    
def ids_submission():
            
    ids1 = ids_marks.get()
    name1 = name.get()
        
    conn = sqlite3.connect("New.db")
    with conn:
        cur = conn.cursor()
#    addColumn2 = "ALTER TABLE DETAILS ADD COLUMN IDS varchar(32)"
#    cur.execute(addColumn2)
    cur.execute( '''UPDATE DETAILS SET IDS == "%s" WHERE Name == "%s"''' %(ids1,name1))
    conn.commit()
    cur.close()    
    conn.close()     
    
'''   -------------      EMFT SUBMISSION     ------------      ''' 
    
def open_window_emft():
    global emft_marks
    global emft_entry
    emft_marks = IntVar()
    
    emft = Toplevel()
    emft.title("EMFT")
    emft.geometry("400x400")
    Label(emft, text = "Enter the marks obtained :").pack()
    
    emft_entry = Entry(emft,textvariable = emft_marks)
    emft_entry.pack()
    
    Button(emft, text = "Done", command = emft_submission).pack()
    
def emft_submission():
            
    emft1 = emft_marks.get()
    name1 = name.get()
        
    conn = sqlite3.connect("New.db")
    with conn:
        cur = conn.cursor()
#    addColumn3 = "ALTER TABLE DETAILS ADD COLUMN EMFT varchar(32)"
#    cur.execute(addColumn3)
    cur.execute( '''UPDATE DETAILS SET EMFT == "%s" WHERE Name == "%s"''' %(emft1,name1))
    conn.commit()
    cur.close()    
    conn.close()
    
 
    
def do():
    open_window()
    put()

'''       ----------       INVALID SIGN IN      ------------  '''    
    
def invalid():
    screen3 = Toplevel(screen)
    screen3.title("Checking")
    screen3.geometry("300x250")
    Label(screen3, text = "Invalid UserID or Password.",font = (9), fg = "red").pack()
    
'''        ---------------          CGPA CALCULATION     -------------        '''

def cgpa_func():
     global cgpa
     fct1 = fct_marks.get()
        
     ids1 = ids_marks.get()
    
     emft1 = emft_marks.get()
        
     avg = 0   
     tot = 0
     avg_per = 0 
     cgpa = 0
    
     tot = fct1 + ids1 + emft1
     avg = tot/300 
     avg_per = avg * 100
     
     cgpa = (avg_per)/9.5
     print("The cgpa is:", cgpa ) 
        
def display_cgpa():
    
    tuf = Toplevel()
    tuf.title("Displaying CGPA")
    tuf.geometry("300x300")
    Label(tuf , text = "THE CGPA OBTAINED IS :",font = ("arial",9,"bold"),bg = "yellow",fg = "red").place(x = 78, y = 40)
    Label(tuf, text= float(cgpa),bg = "cyan").place(x = 88,y = 90)  
        
def do1():
    cgpa_func()
    display_cgpa()  
   
'''      --------------      GRADE CALCULATION     ---------------'''
#       CGPA          GRADE
#     9 - 10.0          O
#     8 - 8.9           E
#     7 - 7.9           A
#     6 - 6.9           B
#     5 - 5.9           C
#     below 5           D
 


def grade_func():
    global grade    
    
    if (cgpa < 5.0):
        grade = "D"
    elif(cgpa < 6.0):
        grade = "C"
    elif(cgpa < 7.0):
        grade = "B"
    elif(cgpa < 8.0):
        grade = "A"
    elif(cgpa < 9.0):
        grade = "E"
    elif(cgpa < 10.1):
        grade = "O"
    
    print("The grade obtained is:", grade)    
        
def display_grade(): 
    tub = Toplevel()
    tub.title("Displaying Grade")
    tub.geometry("300x300")
    
    Label(tub , text = "THE GRADE OBTAINED IS :",font = ("arial",9,"bold"),bg = "yellow",fg = "red").place(x = 78, y = 40)
    Label(tub, text= str(grade),width = 10,bg = "cyan").place(x = 96,y = 90) 
    
def do2():
    grade_func()
    display_grade()  
    
'''         ------------        GUI 3        ---------------        '''

def open_window3():
    
    tom = Toplevel()
    tom.title("GUI 3")
    tom.geometry("500x500")
    tom.configure(bg = "aquamarine")
    
    Label(tom, text = "RESULTS",font = ("arial",14,"bold")).place(x=145,y=40)
    
    Label(tom, text = "Input options :",font = ("arial",11,"bold")).place(x=50,y=90)
    
    Button(tom, text = "CGPA",width = 10,bg = "yellow",command = do1).place(x=70,y=130)
    
    Button(tom, text = "GRADE", width =10,bg = "yellow", command = do2).place(x=70,y=200)

    Button(tom, text = "NEW INPUT", width = 10,bg = "yellow",command = new_input).place(x=250,y=130)
    
    Button(tom, text = "CLOSE", width = 10,bg = "yellow",command = close).place(x=250,y=200)
        
    
'''         -----------------       FINAL CLOSE       -----------------'''
def close():
    screen.destroy()
    
'''          ------------       NEW INPUT        ----------------'''
def new_input():
    
    loggedin()  
    

'''        ------------        DATABASE CONNECTION         ----------       '''    
    
conn = sqlite3.connect('New.db')
cur = conn.cursor()

cur.execute("CREATE TABLE IF NOT EXISTS DETAILS(Name TEXT, Branch TEXT, RegdID REAL)")

#sql = "DELETE FROM DETAILS WHERE Branch = 'wasd'"

#cur.execute(sql)

conn.commit()

#print(cur.rowcount, "record(s) deleted")    
         
#conn.commit()
    

def put():
    name1 = name.get()
    branch1 = branch.get()
    regdid1 = regdid.get()
    conn = sqlite3.connect('New.db')
    cur = conn.cursor()

    cur.execute("INSERT INTO DETAILS (Name,Branch, RegdID) VALUES (?,?,? )",(name1,branch1,regdid1))
    conn.commit()

'''      ----------      SIGN UP CHECKING PART     ----------      '''    

def login_verify():
    user1 = userid.get()
    password1 = password.get()
    username_entry.delete(0, END)
    password_entry.delete(0, END)
    
    list_of_files = os.listdir()
    
    
    if user1 in list_of_files:
        
        file1 = open(user1, "r")
        verify = file1.read().splitlines()
        
        if password1 in verify:
            
            loggedin()
           
        else:
            invalid()
    else:
        invalid()
'''      -----------      OPENING GUI       ------------       '''

def signin():
    global screen
    global userid
    global password
    global username_entry
    global password_entry
    screen = Tk()
    screen.geometry("500x500")
    screen.title("Student Management system")  
    screen.configure(bg = "aquamarine")
    Label(screen, text = "Username :",font = ("arial",10,"bold")).place(x= 50 , y =70 )
    userid = StringVar()
    username_entry = Entry(textvariable = userid)
    username_entry.place(x= 180 , y= 70)
    Label(screen, text = "Password :",font = ("arial",10,"bold")).place(x=50 , y= 130)
    password = StringVar()
    password_entry = Entry(textvariable = password, show = "*")
    password_entry.place(x= 180, y=130)
    Button(screen, text = "Submit",width = 14,fg ='green',font = ("arial",10,"bold"), relief = RAISED ,command = login_verify).place(x=180,y=250)
    screen.mainloop()
   

signin()
cur.close()
conn.close()


'''
                          ----------------     WORKING PROCEDURE    -----------------
            
For signing in, At first , Username and password must be stored in a text file( without .txt format provided name of file
must be the same as username). After successful logging in, GUI(1) appears where basic details are input. Then after 
submitting GUI(1), GUI(2) opens, which gives option for insertion of marks obtained in three subjects. After tapping
Done in respective pages, The marks get saved in their newly assigned columns, near the Name, Branch, RegdId columns
in the database created earlier. Then on hitting submit in GUI(2), GUI(3) appears where four buttons are available.
On tapping CGPA, A new GUI opens within and displays the CGPA obtained. On tapping GRADE, Another new GUI opens within
and displays the Grade obtained as per the specified table (commented) within the code. On tapping NEW INPUT, The code
jumps back to GUI(1), where another new student can input his pariculars and all the process recycles. On tapping CLOSE,
all GUI(s) winds up.

A sample text file is attached within the github link for reference.

'''
```
